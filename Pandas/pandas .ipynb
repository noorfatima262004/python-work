```python


# (merge ,join , concatinate in pandas )
# merge
# The merge() function is used to combine DataFrames based on a common column 
#join
#Types of Joins: You can perform inner, left, right, and outer joins.
#Inner Join: Retains only the rows where the join condition is met.
#Left Join: Retains all rows from the left DataFrame and matching rows from the right DataFrame.
#Right Join: Retains all rows from the right DataFrame and matching rows from the left DataFrame.
#Outer Join: Retains all rows from both DataFrames, filling in missing values with NaN.

# on -> jo uniqe column ho ga us k sath krna ha compare 
# how -> join krny k lia how key word ay ga


# Sample DataFrames
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K3'], 'B': ['B0', 'B1', 'B3']})


```

```python
# Inner Join
inner_merged = pd.merge(left, right, on='key', how='inner')
print(inner_merged)
```

```python
# Left Join
left_merged = pd.merge(left, right, on='key', how='left')
print(left_merged)
```

```python
# right Join
left_merged = pd.merge(left, right, on='key', how='right')
print(left_merged)
```

```python
# Sample DataFrames

#Outer Join: Retains all rows from both DataFrames, filling in missing values with NaN.

left = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K3'], 'B': ['B0', 'B1', 'B3']})

outer_merged = pd.merge(left, right, on='key', how='outer')
print(outer_merged)
```

```python

# jb data same na ho 

left = pd.DataFrame({'key': ['K0', 'K7', 'K2'], 'A': ['A0', 'A1', 'A2']})
right = pd.DataFrame({'key': ['K0', 'K14', 'K13'], 'B': ['B10', 'B1', 'B3']})
inner_merged = pd.merge(left, right, on='key', how='inner')
print(inner_merged)
```

```python

#Concatenating Vertically (Default)

"""
Parameters of pandas.concat()

objs: A sequence or mapping of Series or DataFrame objects to be concatenated.

axis: The axis along which the concatenation will occur. Default is 0 (rows), but you can set it to 1 for columns.

join: Specifies how to handle overlapping labels along the specified axis. Options include 'outer' (default), 'inner', 'left', and 'right'.

ignore_index: If True, it resets the resulting object's index. Default is False.

keys: Optional sequence to add an identifier to the result indexes.

levels: Specific levels to use for constructing a MultiIndex.

names: Names for the levels in the resulting hierarchical index.

verify_integrity: If True, it checks whether the new concatenated axis contains duplicates and raises a ValueError if it does. Default is False.

sort: If True, it sorts the resulting DataFrame or Series by the keys. Default is False.

copy: If False, it avoids copying data unnecessarily. Default is None

"""
# Create two DataFrames
df1 = pd.DataFrame({'letter': ['a', 'b'], 'number': [1, 2]})
df2 = pd.DataFrame({'letter': ['c', 'd'], 'number': [3, 4]})

# Concatenate vertically
result = pd.concat([df1, df2])
print(result)
```

```python

# Create two DataFrames
df1 = pd.DataFrame({'letter': ['a', 'b'], 'number': [1, 2]})
df2 = pd.DataFrame({'animal': ['bird', 'monkey'], 'name': ['polly', 'george']})

# Concatenate horizontally
result = pd.concat([df1, df2])
print(result)
```

```python
# Create two DataFrames
df1 = pd.DataFrame({'letter': ['a', 'b'], 'number': [1, 2]})
df2 = pd.DataFrame({'animal': ['bird', 'monkey'], 'name': ['polly', 'george']})

# Concatenate horizontally
result = pd.concat([df1, df2],axis = 1)
print(result)
```

```python

# Create two DataFrames
df1 = pd.DataFrame({'letter': ['a', 'b'], 'number': [1, 2]})
df3 = pd.DataFrame({'letter': ['c', 'd'], 'number': [3, 4], 'animal': ['cat', 'dog']})

# Concatenate with 'outer' join (default)
result_outer = pd.concat([df1, df3], sort=False)
print(result_outer)

# Concatenate with 'inner' join
result_inner = pd.concat([df1, df3], join="inner")
print(result_inner)
```

```python

# Create two DataFrames
df1 = pd.DataFrame({'letter': ['a', 'b'], 'number': [1, 2]})
df2 = pd.DataFrame({'letter': ['c', 'd'], 'number': [3, 4]})

# Concatenate and reset index
result = pd.concat([df1, df2], ignore_index=True)
print(result)
```

```python
import matplotlib.pyplot as plt
```

```python
x = [20 , 30 , 40 , 50 ]
y = ["PaK" , "ind" , "afg" , "chi"]
ex = [0.2,0.0,0.0,0.0]
cl = ["r" , "b" , "pink" , "gray"]

plt.pie(x , labels= y , explode = ex , colors = cl , shadow = True , radius = 1.3 , autopct = "%0.1f%%" , labeldistance = 1.1 ,
        startangle =50 , textprops = {"fontsize": 15} , wedgeprops = {"edgecolor":"black" , "linewidth":3} )
plt.show()
```

```python
x = [10,20,30,40]
x1 = [12,23,34,45]
x2 = [2,3,4,5]
y = ["A" , "B" , "C" , "D"]

plt.pie(x , labels = y , radius = 1)
plt.pie(x1, labels = y , radius = 0.5)
plt.pie(x2, labels = y , radius = 0.25)
plt.show()
```

```python
x = [10,20,30,40]
x1 = [12,23,34,45]
x2 = [2,3,4,5]
y = ["A" , "B" , "C" , "D"]
cl1 = ["b","g","m","y"]

plt.pie(x , labels = y , radius = 1)
plt.pie(x1,  radius = 0.5)
plt.pie(x2,  radius = 0.2 , colors =cl1)
plt.show()
```

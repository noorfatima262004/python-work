```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
categories = ['A', 'B', 'C', 'D']
values1 = [10, 20, 15, 25]
values2 = [8, 18, 12, 22]

# Create bar plot
bar_width = 0.35
x = np.arange(len(categories))

# Create bars for the first dataset
plt.bar(x - bar_width/2, values1, bar_width, label='Dataset 1', color='skyblue')

# Create bars for the second dataset
plt.bar(x + bar_width/2, values2, bar_width, label='Dataset 2', color='orange' )

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Two Bar Plots in the Same Box')
plt.xticks(x, categories , rotation = 30)

# Add legend
plt.legend(loc = "best")

# Display the plot
plt.show()
```

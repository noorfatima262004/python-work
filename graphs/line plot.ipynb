```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x_1 = [1 , 2 , 3 , 5 , 3]
y_1 = [9 , 8 , 7 , 5 , 6]

# when we use pandas lib 

d= pd.DataFrame({"x_1":x_1,"y_1":y_1})

sns.lineplot(x="x_1",y="y_1" ,  data= d)
plt.show()


```

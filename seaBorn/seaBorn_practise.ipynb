```python
sns.boxplot(y='total_bill', x='day', data=tips)

# Box Plot ek tarah ka graph hai jo data ke distribution ko dikhata hai.
# Box ke beech mein ek line hoti hai jo median (yaani data ka beech wala point) ko dikhati hai.
# Box ke do side mein chhoti lines (whiskers) hoti hain jo data ke minimum aur maximum points ko dikhati hain, lekin kuch extreme values ko chhod kar.
# Agar kisi point ko outlier (bahar ka point) kehte hain, to woh alag se dikhaya jata hai.
# Box Plot yeh batata hai ke data kahan concentrate hai aur kitna spread hai.

# Concentration (Data ka Center):
# Box ke andar line (Median): Yeh line data ka beech wala point dikhati hai. Is se pata chalta hai ke zyada tar data kis value ke around hai.

# Spread (Data ka Phelav):
# Box ki width (Interquartile Range): Box ka size batata hai ke data ka central 50% kitna spread hai. Agar box lamba hai, to data zyada spread hai; agar chhota hai, to data zyada concentrated hai.
# Whiskers: Yeh lines box ke dono side se nikalti hain aur data ke minimum aur maximum values ko dikhati hain (lekin extreme outliers ko nahi). Inse bhi data ke spread ka idea milta hai.

# Spread:
# Meaning: Data ka phelav ya dispersion.
# Example: Agar data points ek wide range me hain (jaise 10 se 100 tak), to data zyada spread hai. Iska matlab hai data ki values ek wide range me scattered hain.
# Box Plot Indicator: Box ki length aur whiskers ki length se pata chalta hai. Agar box aur whiskers lambi hain, to data zyada spread hai.

# Concentration:
# Meaning: Data ka centered ya clustered hona.
# Example: Agar data points ek narrow range me hain (jaise 45 se 50 tak), to data concentrated hai. Iska matlab hai data ki values ek chhoti range me clustered hain.
# Box Plot Indicator: Box ke andar ki line (median) aur box ki width se pata chalta hai. Agar box chhota hai aur median line center me hai, to data concentrated hai.

# Summary:
# Spread: Data ka phelav (wide range) hota hai.
# Concentration: Data ka cluster hona (narrow range) hota hai.
```

```python
sns.jointplot(x='tip', y = 'total_bill', data= tips )
#A joint plot is a type of graph that shows the relationship between two variables.
# It combines a scatter plot (showing how two variables relate) with histograms (showing the distribution of each variable).
# It helps you see both how variables interact and their individual distributions in one view.
```

```python
sns.jointplot(x='tip', y = 'total_bill', data= tips, kind='reg' )
# `kind='reg'` in a joint plot adds a line that shows the trend between two variables.
#  It combines a scatter plot of your data with a line that best fits the points, plus histograms for each variable.
#  This helps you see how the variables are related and their distributions.
```

```python
# Load example data
data = sns.load_dataset("iris")

# Drop non-numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

#data.corr(): Computes the pairwise correlation of columns in the DataFrame.
# Correlation tells us if two things move together (positive correlation) or in opposite directions (negative correlation).
# A value of 1 means they move exactly together.
# A value of -1 means they move in exactly opposite directions.
# A value around 0 means they don't have a strong relationship.
# So, data.corr() calculates these values for each pair of columns in your dataset, helping you understand how they are related.

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# data.select_dtypes(include=['number']) keeps only the numeric columns.
# correlation_matrix is calculated from these numeric columns, and the heatmap is created as before.
```

```python

import pandas as pd

# Load example data
data = sns.load_dataset("iris")

# Convert non-numeric columns to numeric using one-hot encoding
data_encoded = pd.get_dummies(data)

# Calculate the correlation matrix
correlation_matrix = data_encoded.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
# Load example data
data = sns.load_dataset("iris")

# Drop two columns, for example, 'species' and 'sepal_length'
data_dropped = data.drop(columns=['species', 'sepal_length'])

# Create a clustermap with the remaining columns
sns.clustermap(data_dropped.corr(), cmap='coolwarm', annot=True)
```
